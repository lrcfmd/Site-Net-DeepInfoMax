```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
import numpy as np
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
plt.style.use(['science',"no-latex"])
mlp.rc('font', family='sans-serif')
fig, ax = plt.subplots(dpi=400)

df = pd.read_excel("DIM_boxplots.xlsx",header=[0,1,2])

eform_100 = df["Formation Energy"][100]
ax.boxplot([eform_100[i].dropna() for i in eform_100.columns])
ax.set_ylabel("Test MAE (Ev / Atom)")
ax.set(xlabel=None)
ax.set(xticklabels=["Control", "DIM", "Band Gap"])

fig.show()
fig.savefig("eform_100.png",transparent=False,facecolor='white', dpi=400)

```

```python
fig, ax = plt.subplots(dpi=400)

eform_100 = df["Band Gap"][100]
ax.boxplot([eform_100[i].dropna() for i in eform_100.columns])
ax.set_ylabel("Test MAE (eV)")
ax.set(xlabel=None)
ax.set(xticklabels=["Control", "DIM", "Band Gap"])

fig.show()
fig.savefig("egap_100.png",transparent=False,facecolor='white', dpi=400)
```
